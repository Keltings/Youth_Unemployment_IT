```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#ignore warning
import warnings
warnings.filterwarnings('ignore')
```

```python
# a function to read any csv file.
def read_csv_file(file_path):
    dataframe = pd.read_csv(file_path)
    return dataframe

```

```python
#call the function to read the cvs files
kenya = read_csv_file("pop_kenya.csv")
ghana = read_csv_file("pop_ghana.csv")
rwanda = read_csv_file("pop_rwanda.csv")
edu_kenya = read_csv_file("education_kenya.csv")
edu_ghana = read_csv_file("education_ghana.csv")
edu_rwanda = read_csv_file("education_rwanda.csv")
```

```python
#basic analysis
import pandas as pd

def analyze_dataframe(dataframe):
    # Info about the DataFrame
    print("Info:")
    print(dataframe.info())
    print("\n")

    # Description of the DataFrame
    print("Description:")
    print(dataframe.describe())
    print("\n")

    # Check for missing values
    print("Missing Values:")
    print(dataframe.isnull().sum())
    print("\n")

    # Check for duplicates
    duplicates = dataframe.duplicated()
    print("Duplicates:")
    print(duplicates.sum())
    print("\n")

    # Size of the DataFrame
    print("Size:")
    print(dataframe.size)
    print("\n")

    # Shape of the DataFrame
    print("Shape:")
    print(dataframe.shape)
    print("\n")

    # Length of the DataFrame
    print("Length:")
    print(len(dataframe))
    print("\n")
# Call the function to analyze the DataFrame
analyze_dataframe(edu_kenya)
```

```python
# merge the three dataframes of the populations of kenya, rwanda, and ghana


# Then concatenate the data frames
pop_df = pd.concat([kenya, rwanda, ghana], ignore_index=True)
pop_df.head()
# Now, 'merged_df' contains the rows from all three data frames with a 'source' column
```

```python
# Shuffle the rows of the DataFrame
pop_df_shuffled = pop_df.sample(frac=1, random_state=42)
pop_df_shuffled
# The 'frac=1' argument means you want to sample the entire DataFrame
# 'random_state' is set for reproducibility; you can omit it if not needed

# Now, 'shuffled_df' contains the rows of 'merged_df' shuffled randomly
```

```python
analyze_dataframe(pop_df)
```

If we are going to look at the trends in our data then the year column should be in date format and not as an interger and the age and sex columns as categories which is something to note.

```python
# Convert the 'year' column to datetime format
pop_df['year'] = pd.to_datetime(pop_df['year'], format='%Y')

# Convert the 'sex' column to category
pop_df['sex'] = pop_df['sex'].astype('category')
pop_df['age_group'] = pop_df['age_group'].astype('category')

# Now, the 'year' column is of integer type, and the 'sex' column is of categorical type
pop_df.info()
```

IT Employment Trends: Visualize how IT employment has evolved over the years in the "Information and Communication" industry. This can help identify growth patterns and fluctuations.


```python
# Set the style for the Seaborn plots
sns.set(style="whitegrid")
# Create a line plot for trends based on communications, unemployment, and sex
plt.figure(figsize=(12, 6))

# Plot Communications
sns.lineplot(data=pop_df, x='year', y='Information and communication', label='Communications')

# Set the y-axis label
plt.ylabel("Trend Value")
# Show the legend
plt.legend()

# Title and labels
plt.title("Trends of Communications and Unemployment Over Years")
plt.xlabel("Year")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the Seaborn plots
sns.set(style="whitegrid")

# Create a scatter plot for IT employment based on age and differentiated by gender
plt.figure(figsize=(10, 6))
sns.barplot(data=pop_df, x='age_group', y='total_employed_population', hue='sex', palette='Set1')
plt.title("Age and Gender Analysis: Distribution of IT Employment")
plt.xlabel("Age Group")
plt.ylabel("Total Employed Population")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```

```python
# Set the style for the Seaborn plots
sns.set(style="whitegrid")

# Distribution of age_group
plt.figure(figsize=(10, 6))
sns.countplot(data=pop_df, x='age_group')
plt.title("Distribution of Age Groups")
plt.xticks(rotation=45)
plt.show()



```

```python
# Correlation heatmap
corr_matrix = pop_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Box plot of age_group vs. ILO_inactive_share
plt.figure(figsize=(10, 6))
sns.boxplot(data=pop_df, x='age_group', y='ILO_inactive_share')
plt.title("Box Plot: Age Group vs. ILO Inactive Share")
plt.xticks(rotation=45)
plt.show()
```

```python

# Set the style for the Seaborn plots
sns.set(style="whitegrid")

# Univariate analysis - Distribution of ILO_inactive_share
plt.figure(figsize=(10, 6))
sns.histplot(data=pop_df, x='ILO_inactive_share', bins=20, kde=True)
plt.title("Distribution of ILO Inactive Share")
plt.show()

# Bivariate analysis - Scatter plot of ILO_inactive_share vs. total_inactive_population
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pop_df, x='ILO_inactive_share', y='total_inactive_population')
plt.title("Scatter Plot: ILO Inactive Share vs. Total Inactive Population")
plt.show()

# Bivariate analysis - Pairplot for selected columns
selected_columns = ['ILO_inactive_share', 'total_inactive_population', 'ILO_unemployed_rate', 'total_unemployed_population']
sns.pairplot(pop_df[selected_columns])
plt.suptitle("Pairplot of Selected Variables")
plt.show()

# Bivariate analysis - Box plot of age_group vs. total_employed_population
plt.figure(figsize=(10, 6))
sns.boxplot(data=pop_df, x='age_group', y='total_employed_population')
plt.title("Box Plot: Age Group vs. Total Employed Population")
plt.xticks(rotation=45)
plt.show()

# Univariate analysis - Count plot of sex
plt.figure(figsize=(8, 5))
sns.countplot(data=pop_df, x='sex')
plt.title("Count Plot: Gender Distribution")
plt.show()

# Bivariate analysis - Violin plot of age_group vs. total_unemployed_population by sex
plt.figure(figsize=(10, 6))
sns.violinplot(data=pop_df, x='age_group', y='total_unemployed_population', hue='sex', split=True)
plt.title("Violin Plot: Age Group vs. Total Unemployed Population by Gender")
plt.xticks(rotation=45)
plt.show()
```
